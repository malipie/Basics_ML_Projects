```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
import sklearn

sns.set_theme(font_scale=1.3)
np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
```

```python
data, target = make_regression(n_samples=200, n_features=1, noise=20)
target = target ** 2

print(f'{data[:5]}\n')
print(target[:5])
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()
regressor.fit(data, target)

plot_data = np.arange(-3, 3, 0.01).reshape(-1, 1)

plt.figure(figsize=(8,6))
plt.title('Decision tree regression')
plt.plot(plot_data, regressor.predict(plot_data), c='green', label='decision tree regression')
plt.scatter(data, target, label='data')
plt.legend()
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.show()

            
         
```

```python
max_depth_paramether = 1
regressor = DecisionTreeRegressor(max_depth=max_depth_paramether)

regressor.fit(data, target)

plt.figure(figsize=(8,6))
plt.title('Decision tree regression')
plt.plot(plot_data, regressor.predict(plot_data), c='green', label='decision tree regression')
plt.scatter(data, target, label='data')
plt.legend()
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.show()
```

```python
max_depth_paramether = 2
regressor = DecisionTreeRegressor(max_depth=max_depth_paramether)

regressor.fit(data, target)

plt.figure(figsize=(8,6))
plt.title('Decision tree regression')
plt.plot(plot_data, regressor.predict(plot_data), c='green', label='decision tree regression')
plt.scatter(data, target, label='data')
plt.legend()
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.show()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 8))

plot_tree(regressor, 
          feature_names=['feature x'], 
          filled=True, 
          rounded=True)

plt.title("Decision Tree Regressor Visualization")
plt.show()
```
